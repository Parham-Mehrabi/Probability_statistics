### Conditional Probability


- Conditional probability is the likelihood of an event occurring, given that another event has already occurred.
- this is denoted as A given B, expressed as P(A | B), indicating the probability of event A when the event B has already occurred.

to calculate the P(A | B) we have to use the following formula:  $$ P (A | B) = \frac{P(A \cap B)}{P(B)} $$
<img src="images/conditional_b.png" style="float:center; width:35%; background:white"/>



- for example:
    - if the probability of event A is 0.03
    - and the probability of event B is 0.2
    - and the probability of event $ A \cap B $ is 0.01
    - P(A|B) is equal to:
$$ P(A|B) = \frac{P(A \cap B)}{P(B)} = \frac{0.01}{0.2} = \frac{1}{20} = 5\% $$

### Law of total probability
- The law of total probability is a theorem that states, in its discrete case, if $ \{B_n: n = 1, 2, 3, ...\} $ is a
finite or countably infinite set of **mutually exclusive** and **collectively exhaustive** events, 
then for any event A:  
$$ P(A) = \sum_n{ P(A \cap B_n)} $$  
or, alternatively:  
$$ P(A) = \sum_n{P(A|B_n)P(B_n)}$$

### Bayes' theorem
some times we have access to P(A|B) but we are looking for P(B|A), in this cases, we can use bayes' rule to calculate it:  

$$ P(B|A) = \frac{P(A|B)P(B)}{P(A)}$$


<br/>

3. Law of total probability:
    If B1 and B2 are mutually exclusive events such that:
    - $ P(B1 \cap B2) = 0 $ (they can't both happen at the same time), and
    - $ P(B1 \cap B2) = 1 $ (together they cover the whole sample space)
    - for any event A:
        $$ P(A) =  P(A \cap B_1) + P(A \cap B_2) $$
    or
        $$ P(A) = P(A | B1) * P(B1) + P(A | B2) * P(B2)$$

3. Bayes' law:  $$ P(B|A) = \frac{P(A|B)P(B)}{P(A)} $$

### The Cancer problem

Imagine there is a test for a certain type of cancer. The test has a sensitivity of 99% (it correctly detects 99% of people who actually have the cancer) and a specificity of 98% (it correctly identifies 98% of healthy individuals as negative). In a population where, on average, 1 in 10,000 people has this cancer, what is the probability that a person who tests positive actually has the disease?

1. so we have:  
    $ A $ : the person has the cancer  
    $ A^c $ : the person is healthy  
    $ B $ : test result be positive  
    $ B^c $ : test result be negative  

2. what we know is:  
    - $ P(B|A) = .99 $
    - $ P(B^c|A^c) = .98 $
    - $ P(A) = .0001 $

3. and we are looking for $ P(A|B) $  

4. according to Bayes' law:  
    $ P(A|B) = \frac{P(B | A)P(A)}{P(B)} = \frac{.99 \times .0001}{P(B)}$  

5. now we just need to find $ P(B) $  
    - according to the law of total probability:  
        $ P(B) = P(B \cap A) + P(B \cap A^c) $  
    - so we just need to calculate:  
        1. $ P(B \cap A) $
        2. $ P(B \cap A^c) $ 

6. to find $P(B \cap A)$ we can use alternative formula of conditional probability:  
    $ P(B \cap A) = P(B|A)P(A) = .99 \times .0001 = .000099 $

7. now we need to do the same thing to find $ P(B \cap A^c) $ :
    - $ P(B \cap A^c) = P(B|A^c)P(A^c) $
    - $ P(A^c) = 1 - P(A) = 1 - .0001 = .9999$
    - $ P(B|A^c) = 1 - P(B^c|A^c) = 1 - .98 = .02 $
    - $ P(B \cap A^c) = P(B|A^c)P(A^c) = .02 \times 0.9999 = 0.19998$
8. so now we know $ P(B) = .000099 + 0.19998 = 0.020097 $

9. $ P(A|B) = \frac{P(B | A)P(A)}{P(B)} = \frac{.99 \times .0001}{0.020097} \approx .0049 $

10. even though our test has an good accuracy, if it diagnose someone with cancer, the person is more likely to not have cancer
