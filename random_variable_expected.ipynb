### Random Variable

- It is like a function which maps **outcomes** to a digit in $ \mathbb{R} $.  
$ X: \Omega \rightarrow \mathbb{R} $ where $ \Omega$ is sample space

- In other words, a random variable is a function that assigns a real number to each outcome in the sample space  

- There are **discrete** and **continuous** random variables.
    - discrete random variable can link the outcomes of a die for example
    - while a continuous random variable can do it for height of a society

- It is usually shown with a capital letter (like X).

- For example a random variable(X) for a coin can be like this:
    $ X = \begin{cases}
    1 & \text{if the outcome is heads (H)} \\
    0 & \text{if the outcome is tails (T)}
    \end{cases}
    \ $

### PMF

- Probability Mass Function  

- belongs only to discrete

- a probability mass function (sometimes called probability function or frequency function) is a function that gives the probability that a discrete random variable is exactly equal to some value.  

- In other words, this function illustrate the chance of each value of a random variable.  

- For example, imagine X is a discrete random variable, and a is one of its possible values, the out come of X's PMF for a which is shown like $pX(a)$ is equal to the probability of {X = a}:  
    $ pX(a) = P(\{ X = a \}) $  

- if we only have one random variable, we can write the previous notation without like this:  
    $ p(a) = P(\{ X = a \}) $  

- $ p $ is a PMF if it satisfies the following conditions:
    - $ \displaystyle\forall{x} : f(x) \geq 0 $  

    - $ \displaystyle \sum_{x \in \Omega_X} f(x) = 1 $  


- probability X be a part of S, when S is a subset of X's range, is equal to sum of each S value in X for example:  
    - $ Range(x) = \{1, 2, 3, 4, 5, 6\} $
    - $ S = \{2, 3\} \rightarrow S \subseteq R_x $
    - $ \mathbb{P}(X = 2) = \frac{1}{6} $ and $ \mathbb{P}(X = 3) = \frac{1}{6} $
    - $ \mathbb{P}(X \subset \{2, 3\}) = \sum_{x \in S}\mathbb{P}(X=x)$
    - or in other words, chance of X being in S is equal to sum of chance of each S in X's range  

<br/>

**Example**  
imagine we flip 3 coins, and X is a random variable for the number of HEADS for these tosses  
- our sample space would be like this:  
    $ \Omega = \{HHH, HHT, HTH, HTT, THH, THT, TTH, TTT\}$ 
- X would be like this:  
    $
    X = \begin{cases}
    0 & \omega \in \{TTT\}\\
    1 & \omega \in \{HTT, THT, TTH\} \\
    2 & \omega \in \{HHT, HTH, THH\}\\
    3 & \omega \in \{HHH\}\\
    \end{cases}
    $  

- the result of its PMF would be like this:  
$
 p(x) = \begin{cases}
\frac{1}{8} & x = 0 \\
\frac{3}{8} & x = 1 \\
\frac{3}{8} & x = 2 \\
\frac{1}{8} & x = 3 \\
0 & \text{otherwise}
\end{cases} $

### CDF

- Cumulative Distribution Function  

- Defined as:  
    $ \displaystyle F(x) = P(X \leq x)$

- The CDF is obtained by summing the PMF values up to $x$
- Its main features on discrete random variables:  
    - $ \lim_{x\to-\infty}F(x)=0 $  
    - $ \lim_{x\to\infty}F(x)=1 $
    - it is a monotonically non-decreasing function. $ x_1\leq x_2\Rightarrow F(x_1)\le F(x_2) $


- consider the PMF then if you sum up the values for range of X, the result is the result of CDF  

- for example:
    - for a die, the range of x is like this: $ R_x = \{1, 2, 3, 4, 5, 6 \} $
    - the values for each $ p(x) = \frac{1}{6}$
    - so the value of the CDF function would be like this:  
        |x|PMF|CDF|
        |:-:|:-:|:-:|
        |1| $ \frac{1}{6} $ | $ \frac{1}{6} $ |
        |2| $ \frac{1}{6} $ | $ \frac{2}{6} $ |
        |3| $ \frac{1}{6} $ | $ \frac{3}{6} $ |
        |4| $ \frac{1}{6} $ | $ \frac{4}{6} $ |
        |5| $ \frac{1}{6} $ | $ \frac{5}{6} $ |
        |6| $ \frac{1}{6} $ | $ \frac{6}{6} $ |
